```python
class Student:
  '''

  '''

  def __init__(self,id,name,phno,address,Admitted):
    self.Admitted = False
    self.name=name
    self.id=id
    self.phno = phno
    self.address = address
  
  def display(self):
    print(self.id, self.name)
```

```python
class Teachers:

  def __init__(self,id,name,phNo,Address):
     self.id=id
     self.name=name
     self.phNo=phNo
     self.Address=Address
     

  def display(self):
    print(f'Teacher:\n{self.id}  {self.name}')
```

```python
class Subject:
  
  def __init__(self,id,name,classId):

    self.id=id
    self.name=name
    self.classId=classId
    

  def display(self):
    print(f'Subject Id: {self.id} Name: {self.division}')

   
```

```python
class Class:
 
  def __init__(self,id,name,division):

    self.id=id
    self.name=name
    self.division=division
    

  def display(self):
    print(f'Class: {self.id} Division: {self.division}')
```

```python
class Division():
  
  def __init__(self,id,name):
    self.id = id
    self.name = name
    

  def display(self):
    print("Division Information: ", self.name, self.id)

  def displayDivisions():
    print('Department ID          Department Name\n')
    for id, div in SchoolAdmin.Divisions.items():
      print(f' {id}                       {div.name}')
```

```python
from typing import NamedTuple
class SchoolAdmin:
  SchoolAdmin = dict()
  SUBJECTS=dict()
  Divisions =dict()
  Classes = dict()
  Teachers= dict()
  Students = dict()
  
  def __init__(self, id, name, password):
    self.id = id
    self.name =name
    self.password =password
    SchoolAdmin.SchoolAdmin[name] =self
  

  def logout():
    print('You are logged out')

  def AddNewTeacher(id,name,phNo,address):
    if id in SchoolAdmin.Teachers:
      print("This is an exising ID")
    else:
      Teacher = Teachers(id,name,phNo,address)
      SchoolAdmin.Teachers[id]=Teacher
      print(f'{name} is added')

  
  def ModifyTeacher(id,key,val):  
    if key in ['name','phNo','Address']:
      print('here')
      setattr(SchoolAdmin.Teachers[id],key,val)
      print(vars(SchoolAdmin.Teachers[id])) 
      print("You have successfully updated the teacher's profile")
    elif key not in ['name','phNo','Address']:
      print("Teachers class doesn't have this attribute ") 
    else:  
      print("Id not found") 

  def AddNewStudent(id,name,phno,address,Admitted):
     if id in SchoolAdmin.Students:
      print("This is an exising ID")
     else:
      student=Student(id,name,phno,address,Admitted) # create student instance
      SchoolAdmin.Students[id] = student #save the student object
      print(f'{name} is added')


  def ModifyStudent(id,key_,val): 
    if key_ in ['name','phno','address','Admitted']:
      setattr(SchoolAdmin.Students[id],key_,val)
      print(vars(SchoolAdmin.Students[id]))
      print("You have successfully updated the student's profile")
    elif key_ not in ['name','phno','address','Admitted']:
      print("Students class doesn't have this attribute ") 
    else:  
      print("Id not found") 

  
  def AddNewSubject(id,name,classId):   
     if id in SchoolAdmin.SUBJECTS:
      print("This is an exising ID") 
     else:  
      subject=Subject(id,name,classId)        # create subject instance
      SchoolAdmin.SUBJECTS[id] = subject          #save the subject object
      print(f'{name} is added')
    
  
  def ModifySubject(id,key_,val):                            
    if key_ in ['name','classId']:
      setattr(SchoolAdmin.SUBJECTS[id],key_,val)
      print(vars(SchoolAdmin.SUBJECTS[id]))
      print("You have successfully updated the subject's data")
    elif key_ not in ['name','classId']:
      print("subject class doesn't have this attribute")
    else:
      print('Subject Id not found')

  def AddNewClass(id,name,division): 
    if id in SchoolAdmin.Classes:
      print("This is an exising ID")
    else:
      Class_ = Class(id,name,division) #create a class instance
      SchoolAdmin.Classes[id] = Class_
      print(f'{name} class is  successfully added')

  def ModifyClass(id,key_,val): 
    if key_ in ["name","division"]:
      setattr(SchoolAdmin.Classes[id],key_,val)
      print(vars(SchoolAdmin.Classes[id]))
      print("You have successfully updated the class' data")
    elif key_ in ["name","division"]:
      print("Attribute does not exist")
    else:
      print('Id not found')

  def AddNewDivision(id,name):
    if id in SchoolAdmin.Teachers:
      print("This is an exising ID")
    else:
      Division_ = Division(id,name)
      SchoolAdmin.Divisions[id] = Division_
      print(f'{name} division is added')

  def ModifyDivision(id,key_,val): 
    if key_ =='name':
      setattr(SchoolAdmin.Divisions[id],key_,val)
      print(vars(SchoolAdmin.Divisions[id]))
      print("You have successfully updated the division's information")
    elif key_!='name':
      print("Class Division doesn't have this attribute")
    else:
      print('Id not found')
    
  def displaySubjects():
    for sub in SchoolAdmin.SUBJECTS.values():
      print(vars(sub))


  
```
